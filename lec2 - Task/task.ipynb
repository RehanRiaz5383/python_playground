```python
firstNumber = int(input("Please enter the first number: "))
secondNumber = int(input("Please enter second number: "))
primes = []
for number in range(firstNumber,secondNumber+1):
    totalDivisers = 0
    for i in range(1,(number//2)+1):
        if number%i == 0:
            totalDivisers += 1
        if totalDivisers >= 2:
            break
    if totalDivisers <= 1:
        primes.append(number)
print(primes)
    
```
