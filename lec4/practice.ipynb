```python
def isEven(number):
    return number%2 == 0


print(isEven(2))
print(isEven(11))
```

```python
is_odd = lambda number: number%2 != 0

print(is_odd(22))
print(is_odd(11))
```

```python
evens = [x for x in range(20) if (lambda number:number%2 == 0)(x)]
print(evens)
```

```python
natural_numbers = [x for x in range(1,11)]
print(natural_numbers)
squares = list(map(lambda x:x**2,natural_numbers))
print(squares)
```

```python
def squares(number,number2):
    for i in range(number,number2+1):
        yield(i**2)


itr = squares(10,20)


while True:
    try:
        number = next(itr)
        print(number)   
    except Exception as e:
        break
```

```python
def example_function():
    yield "Rehan"
    yield "Faisal"
    yield "Junaid"


data = example_function()
try:
    itr = next(data)
    print(itr)
    itr = next(data)
    print(itr)
    itr = next(data)
    print(itr)
    itr = next(data)
    print(itr)
except Exception as e:
    print("End of incoming")
```
