Task 1: The High-Value Laptops (Filtering & Math)
Ek tech store owner dekhna chahta hai ke uske mehnge laptops ki sale kaisa perform kar rahi hai.

Goal: Data main se aise sabhi records filter karein jahan 'Category' sirf 'Laptop' ho aur unki 'Price' 320,000 PKR se zyada ho.

Action: In premium laptops se hone wali total kamai (Total Revenue) calculate karein.

Hint: Pehle DataFrame ko filter karein, phir ek naya column banayain (maslan Total_Sale = Price * Quantity), aur aakhir main us naye column ka .sum() nikal lein.

```python
import pandas as pd
df = pd.read_csv("sales_data.csv")
majorSoldLaptops = df[(df['Category'] == 'Laptop') & (df['Price'] > 320000)]
majorSoldLaptops['Total_Sale'] =  majorSoldLaptops['Price'] * majorSoldLaptops['Quantity']
totalRevenueGenerated = majorSoldLaptops['Total_Sale'].sum()
print(f"Total Revenue generated through laptops is {totalRevenueGenerated}")
```

Task 2: Top Performing Products (Grouping & Aggregation)
Client ko inventory manage karni hai aur usey pata hona chahiye ke kon sa item sab se zyada bik raha hai aur log payment kaise kar rahe hain.

Action 1: Data ko 'Product' ke hisaab se group karein (groupby use karke) aur har product ki total 'Quantity' sold calculate karein. Pata lagayain ke sab se zyada units kis product ke bike hain.

Action 2: Check karein ke kon sa 'Payment_Method' sab se zyada popular hai (ismein aap value_counts() method try kar sakte hain).

```python
summaryDf = df.groupby("Product").agg({
    'Quantity':'sum',
})
summaryDf = summaryDf.sort_values(by='Quantity', ascending=False)
print(f"The most sold units are of product {summaryDf.index[0]}")
payment_method_used_mostly = df['Payment_Method'].value_counts().idxmax()
payment_counts = df['Payment_Method'].value_counts().iloc[0]
print(f"The payment method: {payment_method_used_mostly}, used {payment_counts} that is most usage in business")
```
