```python
import random

print(random.randint(1, 100))
print(random.choice([1, 2, 3, 4, 5]))
print(random.randrange(1, 100, 2))
print(random.uniform(1, 100))
print(random.random())
print(random.shuffle([1, 2, 3, 4, 5]))
print(random.sample([1, 2, 3, 4, 5], 3))
print(random.choices([1, 2, 3, 4, 5], k=3))
```

```python
import json
data = {"name": "John", "age": 30, "city": "New York"}  #a dictionary
json_data = json.dumps(data) #converting to network json
print(json_data)
print(type(json_data))

parsed_data = json.loads(json_data) #converting to python dictionary
print(parsed_data)
print(type(parsed_data))


```

```python
import csv
with open('data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age', 'City'])
    writer.writerow(['John', 30, 'New York'])
    writer.writerow(['Jane', 25, 'Los Angeles'])
    writer.writerow(['Jim', 35, 'Chicago'])

with open('data.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)




```
