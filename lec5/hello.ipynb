```python
def decorator_function(func):
    def wrapper_function(*args,**kwargs):
        print("Before function call")
        func(*args,**kwargs)
        print("After function call")
    return wrapper_function

@decorator_function
def square(number):
    print(f"The square of {number} is {number**2}")



square(23)
```

```python
def replace_with_string(func):
    func()
    return "Ab main function nahi raha, main ek string hoon!"

@replace_with_string
def my_function():
    print("Main chalna chahta hoon")

# Ab agar aap my_function() call karenge to error aayega
data = my_function() 

# Output: Ab main function nahi raha, main ek string hoon!
```
