```python
#2. switch
alphabet = "a"
match alphabet:
    case 'a' | 'e' | 'i' | 'u' | 'o':
        print('vovel')
    case _:
        print('consonent')
```

```python
#3. loops
print("For loop")
for i in range (1,11):
    print(i)

print("While loop")
x = 1
while x <= 20:
    if x % 2 == 0:
        print(x)
    x = x + 1
```

```python
#4. print line with in place 
print("Hello",end=" ")
print("World")
print("How are you doing today?")
for i in range(1,11):
    if i%3 == 0:
        continue
    print(i, end=" ")
else:
    print("")
    print("Loop ended")
```

```python
#5. itteratable objects
#5(a) lists
friends = ["Rehan","Sanaullah","Subhan","Junaid"]
friends.append("Qasim")
friends.insert(0,"Ali")
unfriended = friends.pop()
print(f"Unfriended: {unfriended}")
for friend in friends:
    print(friend)

print(len(friends))

#list comprehension
numbers = [i if i%2 == 0 else i*i*i for i in range(1,11) ]
print(numbers)

squares = [i**2 for i in range(1,11)]
print(squares)
```
